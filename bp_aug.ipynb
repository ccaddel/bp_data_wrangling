## Aggregate August Raw Data

```python
# import packages
import pandas as pd
import numpy as np
```

```python
# read in csv and pull up first 5 rows
bp_aug = pd.read_csv('aug_raw_data.csv')
bp_aug.head()
```

```python
# pull up dataset info
bp_aug.info()
```

```python
# investigate column values in 'Micro Category'
bp_aug['Micro Category'].unique()
```

```python
# clean up 'None' values by replacing with nans and validate
bp_aug['Micro Category'] = bp_aug['Micro Category'].replace(['None', np.nan])
bp_aug['Micro Category'].unique()
```

#### Are the values in the keyword column (listed below) accurate?

```python
# investigate column values in 'keyword'
bp_aug['Keyword'].unique()
# No misrepresented nans
```

#### Why all the nans in Creative, Clicks, Views, Completed Views, and Conversions?

```python
# amount of nulls in 'Creative','Clicks', 'Views', 'Completed Views','Conversions'
cols = ['Creative','Clicks','Views','Completed Views','Conversions']
for col in cols:
    print(bp_aug[col].isnull().sum())
```

```python
#investigate random columns at the end
list(bp_aug.columns.values)
```

```python
# drop those, and CTR - Subs/Reg Rate and validate
bp_aug = bp_aug.drop(columns=['Unnamed: 38', 'Unnamed: 39', 'Unnamed: 40', 'Unnamed: 41','CTR','CPC','VCR','Bounce Rate',
                             'Reg Rate','  CAC (Reg)','Sub Rate','CAC (Sub)','  Subs/Reg Rate'])
bp_aug.info()
```

### Aggregate data by 'Week Start'

```python
bp_aug_weekly = bp_aug.groupby(['Week Start']).mean()
```
